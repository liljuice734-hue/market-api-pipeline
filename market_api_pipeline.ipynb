```python
import requests
import json
from datetime import datetime

def fetch_and_log_market_rates():
    print("🌐 Connecting to Financial Data API API...")

    # Using a free, public currency exchange rate API
    url = "https://open.er-api.com/v6/latest/USD"

    try:
        response = requests.get(url)

        # Check if connection was successful (HTTP Status 200)
        if response.status_code == 200:
            data = response.json()

            # Extract target exchange rate data (e.g., USD to ZMW, EUR, GBP)
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            rates = data.get("rates", {})

            log_entry = {
                "timestamp": timestamp,
                "base_currency": "USD",
                "target_rates": {
                    "ZMW": rates.get("ZMW"),
                    "EUR": rates.get("EUR"),
                    "GBP": rates.get("GBP"),
                    "CAD": rates.get("CAD")
                }
            }

            # Save the API response to a local JSON log file
            log_filename = "market_rate_log.json"
            with open(log_filename, "w") as f:
                json.dump(log_entry, f, indent=4)

            print("✅ API Connection Successful!")
            print(f"📦 Data successfully logged to: {log_filename}")
            print("\n📊 --- LIVE DATA FETCH PREVIEW ---")
            print(json.dumps(log_entry, indent=4))

        else:
            print(f"❌ API returned an error status code: {response.status_code}")

    except Exception as e:
        print(f"❌ Failed to connect to the API network layer: {str(e)}")

# Execute the live integration script right now
if __name__ == "__main__":
    fetch_and_log_market_rates()
```
